### Single Layer Perceptron

```python
# importing modules
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation

from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
from sklearn.datasets import load_iris
from keras import models
from keras import layers
import pandas as pd

import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
# Now we will convert the pixels into floating-point values.
# Cast the records into float values
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
  
# normalize image pixel values by dividing 
# by 255
gray_scale = 255
x_train /= gray_scale
x_test /= gray_scale
```

```python
#Understand the structure of the dataset
print("Feature matrix:", x_train.shape)
print("Target matrix:", x_test.shape)
print("Feature matrix:", y_train.shape)
print("Target matrix:", y_test.shape)
```

```python
#Visualize the data
fig, ax = plt.subplots(10, 10)
k = 0
for i in range(10):
    for j in range(10):
        ax[i][j].imshow(x_train[k].reshape(28, 28), 
                        aspect='auto')
        k += 1
plt.show()
```

```python
# Form the Input, hidden, and output layers.
model = Sequential([
    
    # reshape 28 row * 28 column data to 28*28 rows
    Flatten(input_shape=(28, 28)),
    
    # output layer
    Dense(10, activation='sigmoid'),  
])
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model
model.fit(x_train, y_train, epochs=10, 
          batch_size=2000, 
          validation_split=0.2)
```

```python
# Find Accuracy of the model
results = model.evaluate(x_test,  y_test, verbose = 0)
print('test loss, test acc:', results)
```

# Fashion MNIST Dataset

```python
(x_train, y_train), (x_test,y_test) = tf.keras.datasets.mnist.load_data()
max_value = 255
x_train / max_value
x_test / max_value
```

# Cross Validation

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train,test_size=0.25, random_state=0)
```

# Define Model

```python
model = Sequential([
    Flatten(input_shape=(28,28,)),
    Dense(512, activation='sigmoid'),
    Dropout(0.5),
    Dense(128, activation='sigmoid'),
    Dropout(0.5),
    Dense(10,activation='softmax'),
])
```

# Compile Model

```python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
x_train.shape, y_train.shape, x_val.shape, y_val.shape, x_test.shape,y_test.shape
```

# Evaluate Model

```python
callbacks = [EarlyStopping(monitor='val_accuracy',patience=3)]
model.fit(x_train,y_train, epochs=100, batch_size=10000, validation_split =0.2, callbacks=callbacks)
```

# Results

```python
results = model.evaluate(x_test, y_test, verbose = 0)
```

```python
print('test_loss, test_acc:', results)
```

**Optimizer that implements the NAdam algorith. Much like Adam is essentially RMSprop with momentum, Nadam is Adam with Nesterov momentum.**

```python
model = Sequential([
    Flatten(input_shape=(28,28,)),
    Dense(512, activation='selu'),
    Dropout(0.6),
    Dense(128, activation='selu'),
    Dropout(0.4),
    Dense(10,activation='softmax'), #output layer with 10 neurons & softmax activation function
])
```

Scaled Exponential Linear Unit (SELU). The Scaled Exponential Linear Unit(SELU) activationfunction is defined as:
    if x > 0: return scale * x
    if x < 0: return scale * alpha * (exp(x)-1)

```python
# Compile Model
model.compile(optimizer='Nadam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#Evaluate Model
callbacks = [EarlyStopping(monitor='val_accuracy',patience=3)]
model.fit(x_train,y_train, epochs=100, batch_size=6000, validation_split =0.2, callbacks=callbacks)
```

```python
results = model.evaluate(x_test, y_test, verbose = 0)
```

```python
print('test_loss, test_acc:', results)
```

# Exponential Linear Unit

The exponential liear unit (ELU) with alpha > 0 is: x if x>0 and alpha*(exp(x)-1) if x<0 The ELU Hyperparameter alpha controls the value to which an ELU saturates for negative net inputs. ELUs diminish the vanishing gradient effect.

```python
model = Sequential([
    Flatten(input_shape=(28,28,)),
    Dense(512, activation='elu'),
    Dropout(0.2),
    Dense(128, activation='elu'),
    Dropout(0.2),
    Dense(10,activation='softmax'), #output layer with 10 neurons & softmax activation function
])
```

```python
# Compile Model
model.compile(optimizer='Nadam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#Evaluate Model
callbacks = [EarlyStopping(monitor='val_accuracy',patience=3)]
model.fit(x_train,y_train, epochs=100, batch_size=6000, validation_split =0.2, callbacks=callbacks)
```

```python
results = model.evaluate(x_test, y_test, verbose = 0)
```

```python
print('test_loss, test_acc:', results)
```
